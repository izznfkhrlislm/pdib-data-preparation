### 2. Deskripsi Data
Dataset HMEQ ini memuat informasi karakteristik dan pelanggaran (*delinquency*) untuk 5960 *home equity loans*. *Home equity loan* adalah sebuah pinjaman dimana *obligor* menggunakan ekuitas dari rumahnya sebagai jaminan. Dataset memiliki karakter sebagai berikut:<br>
- **BAD**: 1 = applicant defaulted on loan or seriously delinquent; 0 = applicant paid loan
- **LOAN**: Amount of the loan request
- **MORTDUE**: Amount due on existing mortgage
- **VALUE**: Value of current property
- **REASON**: DebtCon = debt consolidation; HomeImp = home improvement
- **JOB**: Occupational categories
- **YOJ**: Years at present job
- **DEROG**: Number of major derogatory reports
- **DELINQ**: Number of delinquent credit lines
- **CLAGE**: Age of oldest credit line in months
- **NINQ**: Number of recent credit inquiries
- **CLINO**: Number of credit lines
- **DEBTINC**: Debt-to-income ratio

#### d. Mengecek *Outlier* pada LOAN

```python
import pandas as pd
import matplotlib.pyplot as plt

hmeq = pd.read_csv("hmeq.csv", error_bad_lines=False)
loan = hmeq["LOAN"]

def generate_scatter(df):
    datas = []
    y_pos = range(len(df))
    for data in df:
        datas.append(data)
    plt.scatter(datas, y_pos)
    plt.show()

generate_scatter(loan)
```

**Analisis**: Pada *scatter plot* diatas, terlihat bahwa pada data **LOAN** terdapat *outlier*, yaitu dengan nilai diatas 6000000. Oleh karena itu, saya akan mencari nilai pasti dari *outlier* tersebut.

```python
import pandas as pd

hmeq = pd.read_csv("hmeq.csv", error_bad_lines=False)
outlier = hmeq["LOAN"] > 6000000
res = hmeq[outlier]
res
```

Berdasarkan tabel diatas, *outlier* pada LOAN adalah 6600000. Selanjutnya, saya akan mengubah *outlier* tersebut menjadi *mean* dari keseluruhan data LOAN.

```python
import pandas as pd

hmeq = pd.read_csv("hmeq.csv", error_bad_lines=False)

df_mean = hmeq["LOAN"].mean()

hmeq.loc[hmeq['LOAN'] == 6600000, 'LOAN'] = df_mean

new_df = hmeq["LOAN"]
generate_scatter(new_df)
```

Setelah *outlier* diganti dengan nilai *mean*, data sudah bersih dari nilai *outlier*.
